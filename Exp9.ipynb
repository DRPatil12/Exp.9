```python
# Why NumPy?
## It is faster than lists.
## It is used for numerical computations.
## Base for Pandas
```

```python
#1. Declaration of NumPy:
import numpy as np

a = np.array([1,10,20,30,40,50,500])
b = np.array([[1,2,3],[4,5,6]]) ##ListOfList to print Matrix.
print(a)
print(b)
```

```python
#2. Dimension of Array:
print(a.ndim)
print(b.ndim)
```

```python
#3. Printing (rows,coloumns):
print(a.shape)
print(b.shape)
```

```python
#4. printing datatype: 64bit integer data type
print(a.dtype)
print(b.dtype)
```

```python
#5. In-Built functions using NumPy
print("Zeros=",np.zeros((2,3)))
print("Ones=",np.ones((2,3)))
print("Identity Matrix=",np.eye((3)))
print("Aranged Array=", np.arange(1,10,2))
print("Equal spaced array=", np.linspace(0,1,4))
print("Mean of a = ", np.mean(a)) #float datatype(64bit)
print("Median of a = ", np.median(a)) #float datatype(64bit)
print("max value of a = ", np.max(a))
print("min value of a = ", np.min(a))
print("sum of a = ", np.sum(a))
```

```python
#6. addition and product with a number:
print(a)
print(b)
print(a+5)
print(b*2)
```
